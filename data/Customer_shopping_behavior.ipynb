## Treating Null Values

# Size

```python
df["Size"].unique()
```

```python
df.head()
```

```python
df.groupby("Category")["Size"].unique()
```

```python
df.groupby("Item Purchased")["Size"].unique()
```

```python
df.groupby("Category")["Item Purchased"].unique()
```

```python
df.groupby("Category")["Size"].unique()
```

```python
# Treating all null values or replacing all values in the Electronics categrory.

df.loc[df["Category"] == "Electronics", "Size"] = "Not Applicable"
```

```python
df.groupby("Category")["Size"].unique()
```

```python
# treating all null values or replacing all values in the Accessories categrory.

df.loc[df["Category"] == "Accessories", "Size"] = "Free size"
```

```python
df.groupby("Category")["Size"].unique()
```

```python
# treating all null values or replacing all values in the Footwear categrory.
df.loc[df["Category"] == "Footwear", "Size"] = "Not Applicable"
```

```python
df.groupby("Category")["Size"].unique()
```

```python
# treating all null values or replacing all values in the Clothing categrory.

df.loc[(df["Category"] == "Clothing") & (df["Size"].isnull()), "Size"] = df[df["Category"] == "Clothing"]["Size"].mode()[0]
```

```python
df.isnull().sum()
```

## Review Rating

```python
df.head()
```

```python
df["Review Rating"].mean()
```

```python
df["Review Rating"].median()
```

```python
df["Review Rating"].mode()
```

```python
df.groupby("Item Purchased")["Review Rating"].mean()
```

```python
# Filling null values in the "Review Rating" column with the mean rating for each item purchased
df["Review Rating"] = df["Review Rating"].fillna(df.groupby("Item Purchased")["Review Rating"].transform("mean"))
```

```python
 df.isnull().sum()
```

```python
df.head()
```

```python
df["Review Rating"] = df["Review Rating"].round(2)
```

```python
df.head()
```

# Previous Purchases

```python
# filling previous purchase values with zero
df["Previous Purchases"] = df["Previous Purchases"].fillna(0)
```

```python
df.isnull().sum()
```

# Previous Amount

```python
df.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
df["Purchase Amount (USD)"] = df["Purchase Amount (USD)"].fillna(df.groupby("Item Purchased")["Purchase Amount (USD)"].transform("mean"))
```
